```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt 
from utils.data import load_sample, clean_training_samples

%matplotlib inline
```

```python
samples = clean_training_samples()
```

```python
def display_sample(sample):
    CLASS_ID_COLOURS = {
        '1': (3,  3, 255),
        '2': (3, 255, 3),
        '3': (255, 3, 3),
        '4': (255, 3, 255),
    }
    LABEL_OPACITY=0.15

    image = np.repeat(np.expand_dims(sample[:, :, 0], axis=2), 3, axis=2).astype(np.uint8)
    cv.addWeighted(np.multiply(CLASS_ID_COLOURS['1'], np.repeat(np.expand_dims(sample[:, :, 1], axis=2), 3, axis=2)).astype(np.uint8), LABEL_OPACITY, image, 1.0, 0, image)
    cv.addWeighted(np.multiply(CLASS_ID_COLOURS['2'], np.repeat(np.expand_dims(sample[:, :, 2], axis=2), 3, axis=2)).astype(np.uint8), LABEL_OPACITY, image, 1.0, 0, image)
    cv.addWeighted(np.multiply(CLASS_ID_COLOURS['3'], np.repeat(np.expand_dims(sample[:, :, 3], axis=2), 3, axis=2)).astype(np.uint8), LABEL_OPACITY, image, 1.0, 0, image)
    cv.addWeighted(np.multiply(CLASS_ID_COLOURS['4'], np.repeat(np.expand_dims(sample[:, :, 4], axis=2), 3, axis=2)).astype(np.uint8), LABEL_OPACITY, image, 1.0, 0, image)


    plt.figure(figsize = (30,4))
    plt.imshow(image)
```

```python
sample = samples[samples['has_defect'] == True].sample().iloc[0]
sample = load_sample(sample)
display_sample(sample)
```
